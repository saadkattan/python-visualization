````
import matplotlib.pyplot as plt
import pandas as pd

# To mount your Google Drive.
from google.colab import drive
drive.mount('/content/drive')
````

````
data  = pd.read_csv('/content/drive/My Drive/Colab Notebooks/python-visualization/BlockChain.csv', usecols = ["Date", "Gold Price Index", "Crude Oil WTI", "Bitcoins_in_circulation"], index_col = 0, parse_dates=[0])
data.head()
````

````
plt.subplot(2,2,1)
# Plot the Gold Price Index time series in blue
plt.plot(data["Gold Price Index"],color = "Blue", label = "GPI")
# Plot the Crude Oil WTI time series in green
plt.plot(data["Crude Oil WTI"], color='green', label='COW')
# Add a legend in the top left corner of the plot
plt.legend(loc='upper left')
# Specify the orientation of the xticks
plt.xticks(rotation = 45)
plt.title("GPI and COW")

# Slice Gold Price Index data from '2010' to '2011' inclusive:
data_new = data["Gold Price Index"]["2010":"2011"]

# Plot the sliced data:
plt.subplot(2,2,2)
plt.xticks(rotation=45)
plt.title('Gold Price Index: 2010 to 2011')
plt.plot(data_new, color= 'black')

# Plot March "inset" with the sliced data 
plt.subplot(2,2,3)
plt.xticks(rotation=45)
plt.title('Gold Price Index 2010-2012 with March inset')
plt.plot(data_new, color= 'black')
Marchdata = data_new["2010-03"]
# Specify the axes
plt.axes([0.15, 0.32, 0.2,0.1])
# Plot the inset series in red using the current axes
plt.plot(Marchdata, color = "red")
plt.xticks(rotation=90)
plt.title('March 2010 GPI')


# Improve spacing and display the plot
plt.tight_layout()
plt.show()
````

- Note that: Unlike slicing from standard Python lists, tuples, and strings, when slicing time series by labels (and other pandas Series & DataFrames by labels), the slice includes the right-most portion of the slice.
