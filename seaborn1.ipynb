````
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
````

````
# To mount your Google Drive.
from google.colab import drive
drive.mount('/content/drive')
````

````
# Visualizng Regressions: 

df = pd.read_csv('/content/drive/My Drive/Colab Notebooks/python-visualization/NCHS_-_Death_rates_and_life_expectancy_at_birth.csv')
print(df.columns.values)
sns.lmplot(x="Year" , y="Average Life Expectancy (Years)", data= df)
#print(df.head)
````

````
sns.lmplot(x="Year" , y="Average Life Expectancy (Years)", data= df, hue="Race", palette = "Set1")
````

````
sns.lmplot(x="Year" , y="Average Life Expectancy (Years)", data= df, col = "Sex", hue="Race") #the subplot organized in columns
````

````
sns.lmplot(x="Year" , y="Average Life Expectancy (Years)", data= df, row="Sex") #the subplot organized in rows
````

````
# Generate a black residual plot of the regression between 'Year' and 'Average Life Expectancy (Years)'
sns.residplot(x='Year', y='Average Life Expectancy (Years)', data=df, color='black')
````

````
# Generate a scatter plot of 'weight' and 'mpg' using red circles
plt.scatter(df['Year'], df['Average Life Expectancy (Years)'], label='data', color='red', marker='.')

# Plot in blue a linear regression of order 1 between 'weight' and 'mpg'
sns.regplot(x='Year', y='Average Life Expectancy (Years)', data=df, scatter=None, color='blue', label='order 1') 
# Note that we already plotted the scatter plot in different color > scatter = None. otherwise it will plotted.
# Plot in green a linear regression of order 2 between 'weight' and 'mpg'
sns.regplot(x='Year', y='Average Life Expectancy (Years)', data=df, order=2,scatter=None, color='green', label='order 2')

# Add a legend and display the plot
plt.legend(loc='best')
plt.show()
````
