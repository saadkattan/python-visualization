#### Strip plots:
- Regressions are useful to understand relationships between two continuous variables
- To explore how the distribution of a single continuous variable is affected by a second categorical variable, The strip plot is one way of visualizing this kind of data. It plots the distribution of variables for each category as individual datapoints
- Overlapping points can be difficult to distinguish in strip plots. The argument `jitter=True` helps spread out overlapping points.
- It could be Vertical or horizental


````
# Make a strip plot of 'charges' grouped by 'children'
plt.subplot(2,1,1)
sns.stripplot(x='children', y='charges', data=df, size = 3)
# Make the strip plot again using jitter and a smaller point size = 3, jitter = True to show overlap
plt.subplot(2,1,2)
sns.stripplot(x='children', y='charges', data=df, jitter=True, size = 3)
plt.show()
````

#### Swarm plots: 

-  It spreads out the points to avoid overlap and provides a better visual overview of the data.
- Provides a better alternative to a visually crowded strip plot




````
# Generate a swarm plot of 'charges' grouped horizontally by 'children'  
plt.subplot(2,1,1)
sns.swarmplot(x = 'children', y = 'charges', data = df, size=3)
# Generate the swarm plot where the grouping "children" is on the vertical axis: oreintation is horizontal
plt.subplot(2,1,2)
sns.swarmplot(x = 'charges', y = 'children', data = df , orient = 'h', hue = 'sex', size =3)
# Display the plot
plt.show()

````

#### Violin Plot
Both strip and swarm plots visualize all the datapoints. It's often useful for a larger dataset to use plot types which reduce a dataset to more descriptive statistics and provide a good summary of the data.

````
# Generate a violin plot of 'charges' grouped horizontally by 'children'
plt.subplot(2,1,1)
sns.violinplot(x='children', y='charges', data=df)
# Generate the same violin plot again with a color of 'lightgray' and without inner annotations
plt.subplot(2,1,2)
sns.violinplot(x='children', y='charges', data=df, inner = None, color = 'lightgray')
# Overlay a strip plot on the violin plot
sns.stripplot(x='children', y='charges', data=df, jitter = True, size = 1.5)
# Display the plot
plt.show()
````
