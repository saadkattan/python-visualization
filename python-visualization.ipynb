````

# Import matplotlib.pyplot
import matplotlib.pyplot as plt
# import numpy
import numpy as np
````

````
# generate graphs
x_axes = range(100)
x_1 = np.random.randint(0,100,100) #np.random.randint(min, max, number)
x_2 = np.random.randint(0,100,100)
x_3 = np.random.randint(0,100,100)
x_4 = np.random.randint(0,100,100)
````

````
# Plots on common axes

# Plot in blue x_1 for the x_axes 
plt.plot(x_axes, x_1, color = 'blue') # x_axes, X_1 could be np array, list of pandas series.

# Plot in red x_2 for the x_axes 
plt.plot(x_axes, x_2, color = 'red')

# Display the plot
plt.show()
````

````
# Plot different line plots on distinct axes

# plt.axes([xlo, ylo, width, height]) units are between [0,1]

# Create plot axes for the first line plot
plt.axes([0.05, 0.05, 0.425, 0.9]) 

# Plot in blue x_1 for the x_axes 
plt.plot(x_axes, x_1, color = 'blue')
plt.xlabel("Time")
plt.ylabel("Value")

# Create plot axes for the second line plot
plt.axes([0.525, 0.05, 0.425, 0.9])

# Plot in red x_2 for the x_axes 
plt.plot(x_axes, x_2, color = 'red')

# Display the plot
plt.show()
````

````
#Using subplot:

# Create a figure with 2x2 subplot layout and make the top left subplot active
plt.subplot(2,2,1)

# Plot in blue x_1 for the x_axes 
plt.plot(x_axes, x_1, color = 'blue')
plt.title("x1")

# Make the top right subplot active in the current 2x2 subplot grid 
plt.subplot(2,2,2)

# Plot in red x_2 for the x_axes 
plt.plot(x_axes, x_2, color = 'red')
plt.title('x2')

# Make the bottom left subplot active in the current 2x2 subplot grid
plt.subplot(2,2,3)

## Plot in red x_3 for the x_axes 
plt.plot(x_axes, x_3, color = 'green')
plt.title('x3')

# Make the bottom right subplot active in the current 2x2 subplot grid
plt.subplot(2,2,4)

# Plot in red x_4 for the x_axes 
plt.plot(x_axes, x_4, color = 'black')
plt.title('x4')

# Improve the spacing between subplots and display them
plt.tight_layout() #graphs are cleaner use it for subplot
plt.show()
````

````
# Modifying Axis Limit Individually: plt.xlim([25, 100])

# plotting X4:
plt.plot(x_axes, x_4, color = 'black')
# Set the x-axis range
plt.xlim([25, 100])

# Set the y-axis range
plt.ylim([40, 70])

# Save the image as 'xlim_and_ylim.png' should be located before plt.show()
plt.savefig('X4 with axis limit modification.png') #save the changes on the plot before the instruction only.

# Add a title and display the plot
plt.title('Axis Modification \n X4')
plt.show()
````

````
# Modifying Axis Limit with one indtruction: plt.axis([1990, 2010, 0, 50])

# plotting X2:
plt.plot(x_axes, x_2, color = 'red')

# Set the x-axis and y-axis limits
plt.axis([30, 70, 0, 50])

# Add a title and display the plot
plt.title('Axis Modification \n X2')
plt.show()
````
